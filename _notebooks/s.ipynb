```python
import tkinter as tk

# Your own code goes here


def execute_code():
    user_code = code_text.get("1.0", "end-1c")  # Get code from the text area
    try:
        # Redirect the standard output to our Text widget
        import sys
        original_stdout = sys.stdout
        sys.stdout = TextRedirector(output_text)

        exec(user_code)  # Execute the user's code

        # Restore the standard output
        sys.stdout = original_stdout
    except Exception as e:
        output_text.insert("insert", f"Error: {str(e)}\n")

# Create a custom text widget to redirect stdout
class TextRedirector:
    def __init__(self, text_widget):
        self.text_widget = text_widget

    def write(self, text):
        self.text_widget.insert("end", text)
        self.text_widget.see("end")

# Create the main window
root = tk.Tk()
root.title("Neptune Notebook")

# Create a text area for code input
code_text = tk.Text(root, width=80, height=20)
code_text.pack()

# Create a button to execute the code
execute_button = tk.Button(root, text="Run Code", command=execute_code)
execute_button.pack()

# Create an output console
output_text = tk.Text(root, width=80, height=10)
output_text.pack()

root.mainloop()
```
