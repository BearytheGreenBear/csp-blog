```python
import pygame
import sys
import math

# Initialize Pygame
pygame.init()

# Constants
WIDTH, HEIGHT = 800, 600
CENTER_X, CENTER_Y = WIDTH // 2, HEIGHT // 2
BALL_RADIUS = 20
FPS = 60
WHITE = (255, 255, 255)
FLOOR_SPEED = 2

# Initialize screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Changing Floor Game")

# Clock to control the frame rate
clock = pygame.time.Clock()

# Ball properties
ball_x = CENTER_X
ball_y = CENTER_Y
ball_velocity_y = 0
gravity = 0.5

# Function to calculate flat floor position
def flat_floor(x):
    return CENTER_Y

# Function to calculate sine wave floor position
def sin_wave_floor(x):
    return math.sin(x * 0.02) * 100 + CENTER_Y

# Function to calculate square root floor position
def sqrt_floor(x):
    return math.sqrt(x) * 10 + CENTER_Y

# Function to calculate parabola floor position
def parabola_floor(x):
    return -(x - CENTER_X) ** 2 / 100 + CENTER_Y

# Initial x-position of the floor
floor_x = 0

# Initial floor mode (0: Flat, 1: Sin Wave, 2: Square Root, 3: Parabola)
floor_mode = 0
floor_functions = [flat_floor, sin_wave_floor, sqrt_floor, parabola_floor]

# Function to change the floor type
def change_floor_type():
    global floor_mode
    floor_mode = (floor_mode + 1) % len(floor_functions)

# Main game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                change_floor_type()

    # Clear the screen
    screen.fill((0, 0, 0))

    # Calculate the new ball position
    ball_velocity_y += gravity
    ball_y += ball_velocity_y

    # Update the x-position of the moving floor
    floor_x -= FLOOR_SPEED

    # Calculate the floor position based on the selected mode
    floor_function = floor_functions[floor_mode]

    # Draw the floor
    for x in range(floor_x, WIDTH, 10):
        floor_y = floor_function(x)
        pygame.draw.line(screen, WHITE, (x - floor_x, floor_y), (x + 10 - floor_x, floor_y), 2)

    # Draw the ball
    pygame.draw.circle(screen, WHITE, (ball_x, int(ball_y)), BALL_RADIUS)

    # Check for collision with the floor
    if ball_y + BALL_RADIUS > floor_function(ball_x - floor_x):
        ball_y = floor_function(ball_x - floor_x) - BALL_RADIUS
        ball_velocity_y = -ball_velocity_y * 0.6

    pygame.display.flip()
    clock.tick(FPS)

# Quit Pygame
pygame.quit()
sys.exit()
```
