```python
from PIL import Image, ImageOps
import base64
from io import BytesIO
import IPython.display as display
```

```python
def imageToBase64(image):
    buffered = BytesIO()
    image.save(buffered, format="PNG")
    img_str = base64.b64encode(buffered.getvalue())
    img_str = img_str.decode('utf-8')
    return img_str

def base64toImage(base64string):
    img_str = base64.b64decode(base64string)
    image = Image.open(BytesIO(img_str))
    return image

def grayscale(image):
    return ImageOps.grayscale(image)

def redscale(image):
    r, g, b = image.split()
    r = r.point(lambda i: i * 1)
    g = g.point(lambda i: i * 0)
    b = b.point(lambda i: i * 0)
    return Image.merge('RGB', (r, g, b))

def bluescale(image):
    r, g, b = image.split()
    r = r.point(lambda i: i * 0)
    g = g.point(lambda i: i * 0)
    b = b.point(lambda i: i * 1)
    return Image.merge('RGB', (r, g, b))

def greenscale(image):
    r, g, b = image.split()
    r = r.point(lambda i: i * 0)
    g = g.point(lambda i: i * 1)
    b = b.point(lambda i: i * 0)
    return Image.merge('RGB', (r, g, b))

def colorscale(image, target_color):
    r, g, b = image.split()
    r = r.point(lambda i: i * target_color[0] // 255)
    g = g.point(lambda i: i * target_color[1] // 255)
    b = b.point(lambda i: i * target_color[2] // 255)
    return Image.merge('RGB', (r, g, b))
```

```python
# Sample base64 image string (replace with your actual base64 image string)
base64_image_string = ""

# Convert the base64 string to an image
image = base64toImage(base64_image_string)

# Test grayscale transformation
grayscale_image = grayscale(image)

# Test redscale transformation
red_image = redscale(image)

# Test bluescale transformation
blue_image = bluescale(image)

# Test greenscale transformation
green_image = greenscale(image)

# Test custom color scaling (e.g., scaling to a shade of purple)
target_color = (255, 255, 255)  # Purple color
color_scaled_image = colorscale(image, target_color)

# Create a function to display an image using Pillow
def display_image(img):
    img.show()

# Display the images
display_image(image)
# display_image(grayscale_image)
# display_image(red_image)
# display_image(blue_image)
# display_image(green_image)
display_image(color_scaled_image)
```
